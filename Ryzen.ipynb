```python
import matplotlib.pyplot as plt

# Updated data
keywords = ["Welcome", "NewOrder", "OrderAdd", "OrderComplete", "OrderRemove", "TrackOrder", "TrackOrder(On)", "FallBack","LookOrder","EndChat"]
total_results = [40, 15, 38, 5, 9, 9, 7, 19,30,40]
obtained_results = [40, 13, 36, 5, 6, 8, 7, 19,27,40]
precision = [100, 86.66, 85.71, 100, 66.66, 88.88, 100, 100,80,100]


# Bar chart with updated data
index = range(len(keywords))
width = 0.35  # The width of the bars

plt.figure(figsize=(12, 6))

# Results obtained
plt.bar(index,obtained_results, width, label='Results Obtained', color='skyblue')

# Total results
plt.bar([i + width for i in index], total_results, width, label='Total Results', color='lightgreen')

# Precision
plt.bar([i + 2 * width for i in index], precision, width, label='Precision (%)', color='orange')

plt.xlabel('Keywords')
plt.ylabel('Values')
plt.title('Results Obtained, Total Results, and Precision for Different Keywords')
plt.xticks([i + width for i in index], keywords, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
